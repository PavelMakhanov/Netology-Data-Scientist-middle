```python
google = pd.read_csv('googleplaystore.csv')
```

```python
google.head()
```

```python
pie_data = google['Content Rating'].value_counts()
pie_data.plot(kind='pie', title='Количество приложений по рейтингу контента')
```

**Судя по круговой диаграмме можно предположить, что разработчики в основном делают приложения предназначенные для всеобщего пользования, для того чтобы охватить более широкую аудиторию.**

```python
bar_data = google['Category'].value_counts()
bar_data.plot(kind='bar', title='Количество приложений по категориям')
```

**На диаграмме видно каких категорий приложений больше всего, можем предположить что именно эти категории пользуются большим спросом**

```python
#функция изменяет тип object на int в столбце 'Installs'
def return_int(row):
  if row == 'Free':
     b = 0
  else:
     a =row.replace('+','')
     b = int(a.replace(',',''))
  
  return b
```

```python
google['Installs'].unique()
```

```python
google['Installs>'] = google['Installs'].apply(return_int)
```

```python
google['Price'].unique()
```

```python
#функция изменяет тип object на float в столбце 'Price'
def return_float(row):
  if row == 'Everyone':
     b = 0
  else:
     b =float(row.replace('$',''))
  
  return b
```

```python
google['price'] = google['Price'].apply(return_float)
```

```python
google.head()
```

```python
google.plot(kind='scatter', 
        x='Rating', 
        y='price', 
        title='Корреляция между ценой и оценками пользователей')
```

**По данному графику видно, что низкие оценки в основном у бесплатных и дешевых приложений, но так же у них и много высоких оценок, если зависимость рейтинга от цены есть, но слабая**

```python
google.boxplot(column='Rating')
```

**На данном графике видно что осном распределение оценок лежит в диапозоне от 3 до 5(0.25, 0.75 квартиль). Имеется выброс в виде слишком высокой оценки, скорее всего ошибка в данных, большое количество низких оценок лежит ниже минимального значения, но они в данном случае не являются выбросами**

```python
instals_category= google[['Type',	'Rating','Android Ver']]
```

```python
instals_category = instals_category.pivot_table(index='Android Ver', columns='Type', values='Rating', aggfunc='mean' ).fillna(0)
```

```python
sns.heatmap(instals_category, cmap='PuBu')
```

На данной диаграмме можно сделать вывод, что рейтинг и тип приложения (платный и бесплатный) и версия андроида на оценку не влияют, так же можно сделать вывод, что на некоторые версии андроида платные приложения не выпускались

```python
sns.regplot(data=google, x='Rating', y='Installs>', color='green', marker='+')
plt.title('Корреляция между количеством скачиваний и оценками')
plt.ylim(0, 40)
```

**Судя по данной диаграмме можем предположить, что количество скачиваний на рейтинг не влияет**
