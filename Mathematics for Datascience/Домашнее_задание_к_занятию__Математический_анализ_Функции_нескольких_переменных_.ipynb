Посчитайте частные производные функций:

1) $f(x,y)=2x^2y^3 + 1/x + y^2x + 7$
       
2) $f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y$

```python
import sympy as sym
import numpy as np
```

```python
x = sym.Symbol('x')
y = sym.Symbol('y')
```

```python
prod_x = sym.diff(2*x**2*y**3 + 1/x + y**2*x + 7, x) # частная производная по x
prod_x
```

```python
prod_y = sym.diff(2*x**2*y**3 + 1/x + y**2*x + 7, y) # частная производная по y
prod_y
```
